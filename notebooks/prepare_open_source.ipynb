Посчитаем сколько в среднем пациентов в группе:

```python
from collections import defaultdict
import re
```

```python
def extract_patient(text):
        pattern = r"(patient_\d+)"
        result = re.search(pattern, text)
        if result:
            return result.group(1)
        return None


patients = defaultdict(int)
for path in glob.glob("valid_data/*/*"):
    if (not "ovar" in path) and (not "foll" in path):
        patient = extract_patient(path)
        patients[patient] += 1
```

```python
import numpy as np
```

```python
np.median(np.array(list(patients.values())))
```

```python
cnt = 0
for path in glob.glob("open_source/images/*"):
    cnt += 1
cnt //= 2
cnt
```

```python
patient_ids = np.repeat(np.arange(cnt // 18), 19) + 10000
```

```python
for i, path in enumerate(glob.glob("open_source/images/*")):
    if not "ovar" in path:
        path = Path(path)
        ovar_path = Path(os.path.join(str(path.parent), path.stem + "_ovar.png"))
        os.rename(path, os.path.join(str(path.parent), f"patient_{patient_ids[i]}_" + path.name))
        os.rename(ovar_path, os.path.join(str(ovar_path.parent), f"patient_{patient_ids[i]}_" + ovar_path.name))
```
